```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('anh-xtanh.jpg',0)
def dao_anh(img):
    return 255 - img
negative_transformation = dao_anh(image)

plt.imshow(negative_transformation,cmap='gray')
```

```python
def chuyendoi_log(img,c):
    img = img.astype(float)
    return float(c) * np.log(1 + img)
log_transformation = chuyendoi_log(image,1000)
plt.imshow(log_transformation,cmap='gray')
#https://docs.opencv.org/3.0.0/d3/df2/tutorial_py_basic_ops.html
```

```python
img_task1 = cv2.imread('anh-xtanh.jpg',0)
mean = np.mean(img_task1)
for i in range(img_task1.shape[0]):
    for j in range(img_task1.shape[1]):
        if img_task1[i][j] < mean:
            img_task1[i][j] = 0
        else: 
            img_task1[i][j] = 255
    
plt.imshow(img_task1,cmap='gray')
```

```python
img_task2 = cv2.imread('anh-xtanh.jpg',0)
mean = np.mean(img_task2)
for i in range(img_task2.shape[0]):
    for j in range(img_task2.shape[1]):
        if img_task2[i][j] < mean:
            img_task2[i][j] = 255
        else: 
            img_task2[i][j] = 0
plt.imshow(img_task2,cmap='gray')
```

```python
img_task3 = cv2.imread('anh-xtanh.jpg',0)
mean = np.mean(img_task3)
for i in range(img_task3.shape[0]):
    for j in range(img_task3.shape[1]):
        if (mean - 20 < img_task3[i][j] and img_task3[i][j] < mean + 20):
            img_task3[i][j] = 0
        else: 
            img_task3[i][j] = 255
plt.imshow(img_task3,cmap='gray')
```

```python
url_img = 'anh-xtanh.jpg'
def chuyendoi_gamma(img,c,gamma):
    img = img.astype(float)
    return float(c) * pow(img,float(gamma))
img_1 = cv2.imread(url_img,0)
img_gamma1 = chuyendoi_gamma(img_1,255,0.2)
img_gamma2 = chuyendoi_gamma(img_1,255,0.5)
img_gamma3 = chuyendoi_gamma(img_1,255,1.2)
img_gamma4 = chuyendoi_gamma(img_1,255,1.8)

plt.subplot(1,4,1)
plt.imshow(img_gamma1, cmap='gray')
plt.subplot(1,4,2)
plt.imshow(img_gamma2, cmap='gray')
plt.subplot(1,4,3)
plt.imshow(img_gamma3, cmap='gray')
plt.subplot(1,4,4)
plt.imshow(img_gamma4, cmap='gray')
```

```python
def slicing(img):
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            if(img[i][j] < 100 or img[i][j] > 200):
                img[i][j] = 210
    return img
img_task4 = cv2.imread(url_img,0)
img_task4_ans = slicing(img_task4)
plt.imshow(img_task4_ans,cmap='gray')
# plt.imshow(img_task4)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function to perform gray level slicing
def gray_level_slicing(img, lower_limit, upper_limit, new_value):
    # Create a mask where the values are between lower and upper limits
    mask = (img >= lower_limit) & (img <= upper_limit)
    
    # Set all pixels in the mask to the new value
    img[mask] = new_value
    
    return img

# Load the image (assuming grayscale)
img_task4 = cv2.imread(url_img, 0)

# Apply gray-level slicing
img_task4_ans = gray_level_slicing(img_task4, 100, 200, 210)

# Display the modified image
plt.imshow(img_task4_ans, cmap='gray')
plt.show()
```
