```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# Loading and saving an image
import cv2
path = 'image.png'
gray_img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
cv2.imshow('Grayscale', gray_img)
cv2.waitKey(0)
```

```python
cv2.imwrite('images.png', gray_img)
```

```python
#Converting between color spaces
img = cv2.imread(path)
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imshow('Grayscale image', gray_img)
cv2.waitKey(0)
```

```python
yuv_img = cv2.cvtColor(img, cv2.COLOR_BGR2YUV)
yuv_img
```

```python
cv2.imshow('Y channel', yuv_img[:, :, 0])
cv2.imshow('U channel', yuv_img[:, :, 1])
cv2.imshow('V channel', yuv_img[:, :, 2])
cv2.waitKey(0)
```

```python
image = img[:,:,::-1]
```

```python
plt.imshow(img[:,:,::-1])
```

```python
# Image translation
num_rows, num_cols = img.shape[:2]
translation_matrix = np.float32([ [1,0,70], [0,1,110] ])
img_translation = cv2.warpAffine(img, translation_matrix, (num_cols,
num_rows))
plt.imshow(img_translation)
```

```python
# Dich chuyen anh vao trung tam
num_rows, num_cols = img.shape[:2]
translation_matrix = np.float32([ [1,0,70], [0,1,110] ])
img_translation = cv2.warpAffine(img, translation_matrix, (num_cols +
70, num_rows + 110))
translation_matrix = np.float32([ [1,0,-30], [0,1,-50] ])
img_translation = cv2.warpAffine(img_translation, translation_matrix,
(num_cols + 70 + 30, num_rows + 110 + 50))
plt.imshow(img_translation)
```

```python
# Quay anh
num_rows, num_cols = image.shape[:2]
rotation_matrix = cv2.getRotationMatrix2D((num_cols/2, num_rows/2),
30, 1)
img_rotation = cv2.warpAffine(image, rotation_matrix, (num_cols, num_rows))
# cv2.imshow('Rotation', img_rotation)
# cv2.waitKey(0)
plt.imshow(img_rotation)
```

```python
img_scaled = cv2.resize(image,None,fx=1.2, fy=1.2, interpolation =
cv2.INTER_LINEAR)
# cv2.imshow('Scaling - Linear Interpolation', img_scaled)
img_scaled = cv2.resize(image,None,fx=1.2, fy=1.2, interpolation =
cv2.INTER_CUBIC)
# cv2.imshow('Scaling - Cubic Interpolation', img_scaled)
img_scaled = cv2.resize(image,(450, 400), interpolation = cv2.INTER_AREA)
# cv2.imshow('Scaling - Skewed Size', img_scaled)
# cv2.waitKey(0)
plt.figure(figsize=(10, 5))
plt.subplot(1, 3, 1)
plt.imshow(img_scaled)
plt.title('Scaling - Linear Interpolation')
plt.subplot(1, 3, 2)
plt.imshow(img_scaled)
plt.title('Scaling - Cubic Interpolation')
plt.subplot(1, 3, 3)
plt.imshow(img_scaled)
plt.title('Scaling - Skewed Siz')
plt.show()  
```

```python

rows, cols = img.shape[:2]
#####################
# Vertical wave
img_output = np.zeros(img.shape, dtype=img.dtype)
for i in range(rows):
 for j in range(cols):
     offset_x = int(25.0 * math.sin(2 * 3.14 * i / 180))
     offset_y = 0
     if j+offset_x < rows:
         img_output[i,j] = img[i,(j+offset_x)%cols]
     else:
         img_output[i,j] = 0
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title('Input')
plt.subplot(1, 2, 2)
plt.imshow(img_output)
plt.title('Vertical Wave')
plt.show()   
```
