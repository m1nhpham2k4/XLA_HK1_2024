```python
img2 = img
# plt.imshow(img2)
result = np.zeros_like(img2)
for i in range(img.shape[0] // 2):
    for j in range(img.shape[1]):
        for k in range(img.shape[2]):
            result[i,j,k] = img[i,j,k]
for i in range(img.shape[0] // 2,0, -1):
    for j in range(img.shape[1] - 1,-1,-1):
        for k in range(img.shape[2]):
            result[img.shape[0] - i, j, k] = img[i, j, k]

plt.imshow(result)
```
