```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
#read image
img = cv2.imread('image1.jpg')
iname = "image1.jpg"
plt.imshow(img)
```

```python
# Color image
b,g,r = cv2.split(img) 
imgMerged = cv2.merge((b,g,r)) 
plt.figure() 
plt.subplot(141);plt.imshow(b,cmap='gray');plt.title("Blue channel") 
plt.subplot(142);plt.imshow(g,cmap='gray');plt.title("Green channel") 
plt.subplot(143);plt.imshow(r,cmap='gray');plt.title("Red channel") 
plt.subplot(144);plt.imshow(imgMerged[:,:,::-1]);plt.title("Merged Output")
# cv2.split(img)
```

```python
#Saving an image 
filename = 'sample_img.jpg'
retval = cv2.imwrite(filename, img)

# Check if the image was saved successfully
if retval:
    print(f"Image was successfully saved to {filename}")
else:
    print(f"Failed to save the image to {filename}")
```

```python
# Image properties
height, width, channels = img.shape

# Print properties
print(f'Width: {width}')
print(f'Height: {height}')
print(f'Channels: {channels}')
print(f'Dtype: {img.dtype}')
```

```python
#crop image
crop = img[250:1250, 200:500]
plt.imshow(crop[:,:,::-1])
```

```python
copy = img.copy()
copy[200:300, 50:100] = [0,255,0] #assign green color
plt.imshow(copy[:,:,::-1])
```

```python
#Xoay ảnh
gr = cv2.imread('image1.jpg',0)
result1 = gr.transpose()
plt.imshow(result1, cmap='gray')
```

```python
img_gray = cv2.imread(iname,0)
plt.imshow(img_gray)
```

```python
reslut = img_gray.copy()
reslut = reslut + 50
plt.imshow(reslut,cmap='gray')
```

```python
#flip img
result = img[::-1, :]
plt.imshow(result[:,:,::-1])
```

```python
#Resizing img
h,w  = img.shape[:2]
resized = cv2.resize(img, (200,200))
plt.imshow(resized[:,:,::-1])
```

```python
new_w = 200
aspect_ratio = h/w
new_h = int(new_w*aspect_ratio)
print(new_h)
```

```python
resized = cv2.resize(img,(new_h,new_w))
plt.imshow(resized[:,:,::-1])
```

```python
cv2.imwrite("resized.jpg",resized)
```

```python
img_xam_4bit = (img_gray // 16)
img_xam_2bit = (img_gray // 64)
img_xam_1bit = (img_gray // 128)
plt.figure(figsize=(15,5))
plt.subplot(1,4,1)
plt.imshow(img,cmap='gray')
plt.title('Ảnh xám gốc')
plt.subplot(1,4,2)
plt.imshow(img_xam_4bit,cmap='gray')
plt.title('Ảnh xám 4bit')
plt.subplot(1,4,3)
plt.imshow(img_xam_2bit,cmap='gray')
plt.title('Ảnh xám 2bit')
plt.subplot(1,4,4)
plt.imshow(img_xam_1bit,cmap='gray')
plt.title('Ảnh xám 1bit')
plt.show()
```
