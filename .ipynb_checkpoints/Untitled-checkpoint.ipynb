```python
#read image
img = cv2.imread('image1.jpg')
plt.imshow(img)
```

```python
# Color image
b,g,r = cv2.split(img) 
imgMerged = cv2.merge((b,g,r)) 
plt.figure() 
plt.subplot(141);plt.imshow(b,cmap='gray');plt.title("Blue channel") 
plt.subplot(142);plt.imshow(g,cmap='gray');plt.title("Green channel") 
plt.subplot(143);plt.imshow(r,cmap='gray');plt.title("Red channel") 
plt.subplot(144);plt.imshow(imgMerged[:,:,::-1]);plt.title("Merged Output")
```

```python
#Saving an image 
filename = 'sample_img.jpg'
retval = cv2.imwrite(filename, img)

# Check if the image was saved successfully
if retval:
    print(f"Image was successfully saved to {filename}")
else:
    print(f"Failed to save the image to {filename}")
```

```python
# Image properties
height, width, channels = img.shape

# Print properties
print(f'Width: {width}')
print(f'Height: {height}')
print(f'Channels: {channels}')
print(f'Dtype: {img.dtype}')
```
