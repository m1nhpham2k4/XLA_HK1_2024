Bài 1: Creat a chess table using numpy and opencv

```python
import numpy as np
import cv2

# Define the size of the chessboard
square_size = 50  # Size of each square in pixels
board_size = 8    # Number of squares in each row and column

# Create an 8x8 board using alternating 0s and 1s
board = np.zeros((board_size, board_size), dtype=np.uint8)
board[1::2, 0::2] = 1
board[0::2, 1::2] = 1

# Scale up the board to the desired square size
chessboard = np.kron(board, np.ones((square_size, square_size), dtype=np.uint8))

# Convert the board to a BGR image
chessboard_img = np.zeros((board_size * square_size, board_size * square_size, 3), dtype=np.uint8)
chessboard_img[chessboard == 1] = (255, 255, 255)  # White color for the white squares

# Display the chessboard
cv2.imshow('Chessboard', chessboard_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# Load the images
image1 = cv2.imread('image_8_1.png')
image2 = cv2.imread('image_8_2.png')

# Ensure both images are the same size
if image1.shape != image2.shape:
    image2 = cv2.resize(image2, (image1.shape[1], image1.shape[0]))

# Ensure both images have the same number of channels
if image1.shape[2] != image2.shape[2]:
    if len(image1.shape) == 2:  # If image1 is grayscale
        image1 = cv2.cvtColor(image1, cv2.COLOR_GRAY2BGR)
    if len(image2.shape) == 2:  # If image2 is grayscale
        image2 = cv2.cvtColor(image2, cv2.COLOR_GRAY2BGR)

# Subtract the images
difference = cv2.absdiff(image1, image2)

# Convert the difference image to grayscale (optional)
difference_gray = cv2.cvtColor(difference, cv2.COLOR_BGR2GRAY)

# Apply a binary threshold to emphasize the differences (optional)
_, thresholded_difference = cv2.threshold(difference_gray, 30, 255, cv2.THRESH_BINARY)

# Display the results
cv2.imshow('Difference', difference)
cv2.imshow('Thresholded Difference', thresholded_difference)

# Save the result (optional)
cv2.imwrite('difference.jpg', difference)
cv2.imwrite('thresholded_difference.jpg', thresholded_difference)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# Load the images
image1 = cv2.imread('image_8_1.png')
image2 = cv2.imread('image_8_2.png')

# Get the dimensions of both images
height1, width1 = image1.shape[:2]
height2, width2 = image2.shape[:2]

# Determine the smallest dimensions
min_height = min(height1, height2)
min_width = min(width1, width2)

# Resize both images to the smallest dimensions
image1_resized = cv2.resize(image1, (min_width, min_height))
image2_resized = cv2.resize(image2, (min_width, min_height))

# Proceed with further processing, e.g., subtracting the images
difference = cv2.absdiff(image1_resized, image2_resized)

# Convert the difference image to grayscale (optional)
difference_gray = cv2.cvtColor(difference, cv2.COLOR_BGR2GRAY)

# Apply a binary threshold to emphasize the differences (optional)
_, thresholded_difference = cv2.threshold(difference_gray, 30, 255, cv2.THRESH_BINARY)

# Display the results
cv2.imshow('Difference', difference)
cv2.imshow('Thresholded Difference', thresholded_difference)

# Save the result (optional)
cv2.imwrite('difference.jpg', difference)
cv2.imwrite('thresholded_difference.jpg', thresholded_difference)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
